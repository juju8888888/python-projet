```python
x=np.linspace(-2,2,101)
y=x**2
print(x)
y2=x**4

plt.plot(x,y, label="x^2") #Vectors x, y
plt.title("Graph of x^2 vs x")
plt.plot(x,y2, label="x^4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
x=np.linspace(0, 3*np.pi, 500)

plt.plot(x, np.sin(x**2))
plt.title("A simple chirp")
```

```python
n=int(input("How many points do you want? "))
x=np.linspace(-1,1,n)
plt.plot(x, np.cos(2*np.pi*x))
plt.title("Body function (2pix)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig("cos2pix.png")
plt.show()
```

```python
n=int(input("How many points do you want? "))
x=np.linspace(-1,1,n)
plt.plot(x, np.cos(2*np.pi*x), label="cos(2pix)")
plt.plot(x, np.sin(2*np.pi*x), label="sin(2pix)")
plt.title("Body function (2pix)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.savefig("2pix.png")
plt.show()
```

```python
n=int(input("How many points do you want? "))
x=np.linspace(0,4,n)
y=np.sin(np.pi*x)*np.sin(20*np.pi*x)*np.exp(-x)
plt.plot(x, y)
plt.title("Body function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.savefig("tri.png")
plt.show()
```

```python
n=int(input("How many points do you want? "))
t=int(input("What is the temperature in Kelvin? "))
x=np.linspace(2,10,n)
y=(0.08206*t)/x
plt.xlabel("Molar Volume (L/mol)")
plt.ylabel("Pressure (atm)")
plt.title("Isotherm (ideal gas)")
plt.plot(x,y)
plt.savefig("isotherm.jpg")
plt.show()
```

```python
n=int(input("How many points do you want? "))
h=int(input("How many different temperatures will you take? "))
x=np.linspace(2,10,n)
for i in range(h):
  t=int(input("What is the temperature in Kelvin? "))
  y=(0.08206*t)/x
  plt.plot(x,y, label=("T=",t))
plt.xlabel("Molar Volume (L/mol)")
plt.ylabel("Pressure (atm)")
plt.title("Isotherm (ideal gas)")
plt.legend()
plt.savefig("isotherms.jpg")
plt.show()
```

```python
nb=int(input("Number of points to draw: "))
ini=int(input("Initial value: "))
fin=int(input("Final value: "))
x0=float(input("x0: "))
S=float(input("S: "))
x=np.linspace(ini, fin, nb)
y=(1/np.sqrt(2*np.pi))*np.exp((-1/2)*((x-x0)**2)/(S**2))
plt.plot(x,y)
plt.title("Gaussian Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
nb=int(input("Number of points to draw: "))
x=np.linspace(0,3,nb)
y=np.exp(-x)
plt.plot(x,y,label="e-x")
y2=np.sin(3*np.pi*x)*np.exp(-x)
plt.plot(x,y2,label="sin(3pix)e-x")
plt.title("Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
nb=int(input("Number of points to draw: "))
fu=int(input("Number of functions to draw: "))
ini=int(input("Initial value: "))
fin=int(input("Final value: "))
x=np.linspace(ini,fin,nb)
for i in range(fu):
  x0=float(input("x0: "))
  S=float(input("S: "))
  y=(1/np.sqrt(2*np.pi))*np.exp((-1/2)*((x-x0)**2)/(S**2))
  plt.plot(x,y,label=("xo=",x0,"s=",S))
plt.title("Gaussian functions")
plt.legend()
plt.show()
```
