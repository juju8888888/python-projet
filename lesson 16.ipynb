```python
drinks = pd.read_csv('/content/drinks.csv')
```

```python
drinks.head(10)
```

```python
beer_average = drinks['beer_servings'].mean()
```

```python
beer_average
```

```python
continent_beer_mean = drinks.groupby('continent')['beer_servings'].mean().idxmax()
print("Continent qui boit le plus de bire en moyenne :", continent_beer_mean)
```

```python
wine_statistics = drinks.groupby('continent')['wine_servings'].describe()
print("Statistiques de consommation de vin par continent:\n", wine_statistics)
```

```python
alcohol_mean = drinks.groupby('continent').mean()
print("Moyenne de la consommation d'alcool par continent pour chaque colonne:\n", alcohol_mean)
```

```python
alcohol_median = drinks.groupby('continent').median()
print("Mdiane de la consommation d'alcool par continent pour chaque colonne:\n", alcohol_median)
```

```python
spirit_stats = drinks['spirit_servings'].agg(['mean', 'min', 'max']).reset_index()
spirit_stats.rename(columns={'index': 'Statistique', 'spirit_servings': 'Spirit Consumption'}, inplace=True)
print("Moyenne, minimum et maximum de la consommation d'alcool (spirits) :\n", spirit_stats)
```
