```python
import pandas as pd
baby_names = pd.read_csv("/content/US_Baby_Names_right.csv")
```

```python
# See the first 10 entries
print(baby_names.head(10))
```

```python
# Delete the column 'Unnamed: 0' and 'Id'
baby_names = baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
baby_names.head(10)
```

```python
# Check if there are more male or female names
gender_counts = baby_names['Gender'].value_counts()
print("More male names" if gender_counts['M'] > gender_counts['F'] else "More female names")
```

```python
# Group the dataset by name and assign to names
del baby_names["Year"]
names = baby_names.groupby('Name').sum()
names.head()
```

```python
# How many different names exist in the dataset?
num_unique_names = baby_names['Name'].nunique()
print(f"There are {num_unique_names} different names in the dataset.")
```

```python
# What is the name with the most occurrences?
most_occurred_name = names.size().idxmax()
print(f"The name with the most occurrences is: {most_occurred_name}")
```

```python
# How many different names have the least occurrences?
least_occurred_names = names.size().idxmin()
num_least_occurred_names = names.size().min()
print(f"There are {num_least_occurred_names} different names with the least occurrences, e.g., {least_occurred_names}.")
```

```python
# What is the median name occurrence?
median_occurrence = names.size().median()
print(f"The median name occurrence is: {median_occurrence}")
```

```python
# What is the standard deviation of names?
std_dev_occurrence = names.size().std()
print(f"The standard deviation of name occurrences is: {std_dev_occurrence}")
```

```python
# Get a summary with the mean, min, max, std, and quartiles.
summary = names.size().describe()
print("\nSummary Statistics:")
print(summary)
```

```python
shooting_accuracy = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
shooting_accuracy
```

```python
# Calculate the highest passing accuracy
highest_passing_accuracy = euro12['Passing Accuracy'].max()

# Filter the dataset to exclude goals scored from penalties
goals_without_penalties = euro12[euro12['Penalty goals'] == 0]['Goals']

# Calculate the total goals scored without penalties
total_goals_without_penalties = goals_without_penalties.sum()

# Display the results
print("Highest Passing Accuracy:", highest_passing_accuracy)
print("Total Goals Scored Without Penalties:", total_goals_without_penalties)
```

```python
# Find the team with the highest passing accuracy
team_highest_passing_accuracy = euro12[euro12['Passing Accuracy'] == euro12['Passing Accuracy'].max()]['Team'].values[0]

# Filter the dataset to exclude goals scored from penalties for the team with the highest passing accuracy
goals_without_penalties_highest_accuracy = euro12[(euro12['Penalty goals'] == 0) & (euro12['Team'] == team_highest_passing_accuracy)]['Goals'].values[0]

# Display the results
print("Team with the Highest Passing Accuracy:", team_highest_passing_accuracy)
print("Goals Scored Without Penalties by the Highest Passing Accuracy Team:", goals_without_penalties_highest_accuracy)
```
